```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from scipy.linalg import pinv

class TwinSVC:
    def __init__(self, C1=1, C2=1):
        self.C1 = C1
        self.C2 = C2
        self.w1, self.b1 = None, None
        self.w2, self.b2 = None, None
    
    def fit(self, X):
        """ Entraînement du modèle TSVC """
        # Séparation en deux groupes avec K-Means
        from sklearn.cluster import KMeans
        labels = KMeans(n_clusters=2, random_state=42).fit_predict(X)
        
        X1, X2 = X[labels == 0], X[labels == 1]
        
        # Calcul des hyperplans
        self.w1, self.b1 = self._compute_hyperplane(X1, X2, self.C1)
        self.w2, self.b2 = self._compute_hyperplane(X2, X1, self.C2)
        
        return labels

    def _compute_hyperplane(self, X_pos, X_neg, C):
        """ Résolution de l'équation matricielle pour obtenir les hyperplans """
        m, n = X_pos.shape
        I = np.eye(n)
        
        # Résolution de la minimisation quadratique
        A = X_pos.T @ X_pos + C * I
        B = -np.ones((m, 1))
        
        w = pinv(A) @ X_pos.T @ B  # Pseudo-inverse pour résoudre l'équation
        b = np.mean(X_neg @ w)  # Décalage du biais
        
        return w, b

    def predict(self, X):
        """ Prédiction des clusters """
        d1 = np.abs(X @ self.w1 + self.b1)
        d2 = np.abs(X @ self.w2 + self.b2)
        return np.where(d1 < d2, 0, 1)

# -------------------------------
```

Pour ajouter des frontières de décision entre les clusters dans la visualisation, nous devons afficher les lignes qui séparent les différentes régions où chaque classe est prédite. Cela peut être accompli en traçant des courbes ou des frontières qui représentent les zones où le modèle change de prédiction (en fonction des hyperplans)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

class TwinSVC:
    def __init__(self, C1=1, C2=1):
        self.C1 = C1
        self.C2 = C2
        self.w1, self.b1 = None, None
        self.w2, self.b2 = None, None
    
    def fit(self, X):
        """ Entraînement du modèle TSVC """
        # KMeans pour la séparation des données en 2 clusters
        labels = KMeans(n_clusters=2, random_state=42).fit_predict(X)
        
        X1, X2 = X[labels == 0], X[labels == 1]
        # Calcul des hyperplans pour chaque cluster
        self.w1, self.b1 = self._compute_hyperplane(X1, X2, self.C1)
        self.w2, self.b2 = self._compute_hyperplane(X2, X1, self.C2)
        
        return X1, X2, labels
    
    def _compute_hyperplane(self, X_pos, X_neg, C):
        """ Résolution de l'équation matricielle pour obtenir les hyperplans """
        m, n = X_pos.shape
        I = np.eye(n)
        
        # Résolution de la minimisation quadratique
        A = X_pos.T @ X_pos + C * I
        B = -np.ones((m, 1))
        
        w = np.linalg.pinv(A) @ X_pos.T @ B  # Pseudo-inverse pour résoudre l'équation
        b = np.mean(X_neg @ w)  # Décalage du biais
        
        return w, b
    
    def predict(self, X):
        """ Prédiction des clusters """
        d1 = np.abs(X @ self.w1 + self.b1)
        d2 = np.abs(X @ self.w2 + self.b2)
        return np.where(d1 < d2, 0, 1)

# Entraînement du modèle TSVC
tsvc = TwinSVC(C1=1, C2=1)
X_scaled = np.random.randn(100, 2)  # Exemple de données à 2 dimensions (X_scaled à remplacer par vos données réelles)
X1, X2, labels = tsvc.fit(X_scaled)

# Création d'une grille de points pour dessiner les frontières
x_min, x_max = X_scaled[:, 0].min() - 1, X_scaled[:, 0].max() + 1
y_min, y_max = X_scaled[:, 1].min() - 1, X_scaled[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
Z = tsvc.predict(np.c_[xx.ravel(), yy.ravel()])  # Prédire pour chaque point de la grille
Z = Z.reshape(xx.shape)  # Reshaper la prédiction pour correspondre à la forme de la grille

# Visualisation des clusters et des frontières de décision
plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.4, cmap="Set1")  # Dessiner les zones des clusters
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=labels, palette="Set1", edgecolor='k')
plt.title("Clustering avec TSVC et Frontières de Décision")
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend(title="Cluster")
plt.show()
```
